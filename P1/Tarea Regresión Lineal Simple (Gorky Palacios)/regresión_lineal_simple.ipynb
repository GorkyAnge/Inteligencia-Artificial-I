#### Importar Librerias

```python
import matplotlib.pyplot as plt #Para graficar y visualizar los datos
import pandas as pd #Para manipular y analizar los datos
from sklearn.model_selection import train_test_split #Para dividir los datos en conjuntos de entrenamiento y prueba
from sklearn.linear_model import LinearRegression #Para crear el modelo de regresión lineal 
```

#### Importar Dataset y dividirlo en datos de entrenamiento y pruebas

```python
dataset = pd.read_csv('CO2 Emissions_Canada.csv') #Cargar el conjunto de datos
#X = dataset.iloc[:, 1:3].values #Seleccionar las columnas de características (X) del conjunto de datos
#y = dataset.iloc[:, 3].values #Seleccionar la columna de etiquetas (y) del conjunto de datos

#Tiene doble corchete para que X sea un DataFrame y no una Serie
X = dataset[['Engine Size(L)']].values #Seleccionar la columna de características (X)

y = dataset['CO2 Emissions(g/km)'].values #Seleccionar la columna de etiquetas (y) del conjunto de datos


#Dividir el conjunto de datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1/3, random_state=0)
```

#### Entrenamiento

```python
regressor = LinearRegression() #Crear un objeto de regresión lineal
regressor.fit(X_train, y_train) #Entrenar el modelo con los datos de entrenamiento
y_pred = regressor.predict(X_test) #Predecir los salarios para el conjunto de prueba
```

#### Visualización resultados conjunto de entrenamiento

```python
plt.scatter(X_train, y_train, color='red') #Graficar los datos de entrenamiento
plt.plot(X_train, regressor.predict(X_train), color='blue') #Graficar la línea de regresión para los datos de entrenamiento
plt.title('Engine Size vs CO2 Emissions (Training set)') #Título del gráfico
plt.xlabel('Engine Size') #Etiqueta del eje X
plt.ylabel('CO2 Emissions') #Etiqueta del eje Y
plt.show() #Mostrar el gráfico
```

#### Visualización resultados conjunto de prueba

```python
plt.scatter(X_test, y_test, color='red') #Graficar los datos de prueba
plt.plot(X_test, y_pred, color='blue') #Graficar la línea de regresión para los datos de prueba
plt.title('Engine Size vs CO2 Emissions (Test set)') #Título del gráfico
plt.xlabel('Engine Size') #Etiqueta del eje X
plt.ylabel('CO2 Emissions') #Etiqueta del eje Y
plt.show() #Mostrar el gráfico
```
